```python
import torch
import simulated_bifurcation as sb

# Define the numerical matrix Q, vector l, and constant c
Q = torch.tensor([
    [1.0,  0.5,  0.0],
    [0.5, -1.0, -1.0],
    [0.0, -1.0,  2.0]
], dtype=torch.float32)

l = torch.tensor([-1.0, 0.0, 2.0], dtype=torch.float32)
c = 2.0

poly_model = sb.build_model(Q, vector=l, constant=c, input_type="binary")

# Run the optimization (minimize in this case)
all_vectors, all_values = poly_model.minimize(agents=64, best_only=False)

print("All solutions shape:", all_vectors.shape)
print("All evaluations shape:", all_values.shape)
```

```python
import numpy as np
from scipy.optimize import minimize
from sympy import symbols, exp, lambdify
from qhdopt.qhd import QHD  # Import your QHD implementation

# Step 1: Define the function
x, y = symbols("x y")
f = y**1.5 - exp(4*x) * (y - 0.75)

# Convert function to numerical form
f_numeric = lambdify((x, y), f, "numpy")


# Step 3: Solve with Classical Optimization (Scipy)
def objective(X):
    return f_numeric(X[0], X[1])

# Initial guess (middle of the domain)
x0 = np.array([0.5, 0.5])

# Set bounds for (x, y) in (0,1)
bounds = [(0, 1), (0, 1)]

# Optimize using L-BFGS-B (good for bounded problems)
result = minimize(objective, x0, bounds=bounds, method="L-BFGS-B")


result
```
